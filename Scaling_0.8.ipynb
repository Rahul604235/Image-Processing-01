# Scale the given Image (cells_scale.png) by 0.8 factors.

```python
# Numpy, Pillow, Imageio libraries were used to do the following program. 
import numpy as np
from PIL import Image
import imageio as mp
```

```python
#Orginal Image was opened using pillow library
Original_Image=Image.open('cells_scale.png')
```

```python
Original_Image
```

```python
#Dimension of the original image of Cells
Original_Image.size
```

```python
#Converts image into Numpy array
Source_Matrix=mp.imread('cells_scale.png')
```

```python
Source_Matrix
```

```python
#M,N siginifies rows and columns of the Source_Matrix using shape function
M,N=Source_Matrix.shape
```

```python
#Zero matrix of M*N is created and assigned as Target_Matrix 
Target_Matrix=np.zeros((M,N))
```

```python
#parameters
x=y=0.8
```

```python
#Each element in the Target_Matrix is assigned the value from the Source_Matrix using Bilinear Interpolation method .
for xt in range(0,M):
    for yt in range(0,N):
        xs=xt/x
        ys=yt/y
        a=ys-int(ys)
        b=xs-int(xs)
        z=0
        if 0<=xs<(M-1) and 0<=ys<(N-1):
            Target_Matrix[xt,yt]=(1-a)*(1-b)*Source_Matrix[int(xs),int(ys)]+(1-a)*(b)*Source_Matrix[int(xs),int(ys)+1]+(a)*(1-b)*Source_Matrix[int(xs)+1,int(ys)]+(a)*(b)*Source_Matrix[int(xs)+1,int(ys)+1]
            
        else:
            Target_Matrix[xt,yt]=0
        
```

```python
#Converts the Target_Matrix to the Image and image is saved as 'cells_scale_edited_0.8.png' 
image=Image.fromarray(Target_Matrix)
if image.mode == "F":
    image = image.convert('RGB') 
image.save('cells_scale_edited_0.8.png')  
```
